```python
import zipfile
import os

# ZIP 파일 경로와 추출할 디렉토리 경로 정의
zip_file_path = "/content/drive/MyDrive/278.국내 여행로그 데이터(동부권)/01-1.정식개방데이터/Training/01.원천데이터/TS_csv.zip"
extract_to_dir = "/content/extracted_files"

# 디렉토리가 존재하지 않으면 생성
os.makedirs(extract_to_dir, exist_ok=True)

# ZIP 파일 추출
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    zip_ref.extractall(extract_to_dir)

# 추출된 디렉토리 내 파일 목록 출력
extracted_files = os.listdir(extract_to_dir)
print(extracted_files)
```
