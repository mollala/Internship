```python
#ts : 변경할 파일, ts2 : 참조 파일

# NaN 값을 빈 문자열로 대체하여 문자열을 결합
ts2['SGG_CD_5'] = (ts2['SIDO_NM'].fillna('') +
                   ts2['SGG_NM'].fillna('') +
                   ts2['DONG_NM'].fillna('') +
                   ts2['RI_NM'].fillna(''))

# 결합된 문자열에서 'nan'을 빈 문자열로 대체
ts2['SGG_CD_5'] = ts2['SGG_CD_5'].str.replace('nan', '')

# SGG_CD 열을 문자열로 변환하여 비교
def replace_with_sido_nm(residence_sgg_cd):
    if pd.isna(residence_sgg_cd):  # NaN 값 처리
        return residence_sgg_cd
    try:
        residence_sgg_cd_str = str(int(float(residence_sgg_cd)))  # float 값을 int로 변환 후 문자열로 변환
    except ValueError:
        return residence_sgg_cd  # 변환 불가능한 경우 원래 값을 반환
    match = ts2[ts2['SGG_CD'].astype(str) == residence_sgg_cd_str]
    if not match.empty:
        return match['SGG_CD_5'].values[0]
    return residence_sgg_cd

# RESIDENCE_SGG_CD 열의 값을 업데이트
ts['SGG_CD'] = ts['SGG_CD'].apply(replace_with_sido_nm)

```
