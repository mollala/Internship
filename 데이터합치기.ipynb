```python
import pandas as pd

# CSV 파일을 불러오기
df = pd.read_csv("C:/연구인턴십/하나로만든거/여행이랑_여행객.csv", encoding='utf-8')
activity_df = pd.read_csv('C:/연구인턴십/합친거/tn_activity_consume_his_활동소비내역.csv', encoding='utf-8')
com_df = pd.read_csv('C:/연구인턴십/합친거/tn_companion_info_동반자정보.csv', encoding='utf-8')
mvmn_df = pd.read_csv('C:/연구인턴십/합친거/tn_mvmn_consume_his_이동수단소비내역.csv', encoding='utf-8')
lodge_df = pd.read_csv('C:/연구인턴십/합친거/tn_lodge_consume_his_숙박소비내역.csv', encoding='utf-8')

# 1단계: activity와 adv 데이터 병합
merged_df = activity_df.merge(com_df, on="TRAVEL_ID", suffixes=('_activity', '_com'), how="outer")

# 2단계: 1단계 결과와 lodge 데이터 병합
merged_df = merged_df.merge(lodge_df, on="TRAVEL_ID", suffixes=('', '_lodge'), how="outer")

# 3단계: 2단계 결과와 mvmn 데이터 병합
merged_df = merged_df.merge(mvmn_df, on="TRAVEL_ID", suffixes=('', '_mvmn'), how="outer")

merged_df = merged_df.merge(df, on = "TRAVEL_ID", suffixes=('', '_df'), how="outer")
# 병합된 결과 확인
print(merged_df.head())
```

```python
# 필요한 열 이름을 리스트로 정의합니다.
selected_columns = [
    'TRAVEL_ID', 'VISIT_AREA_ID', 'ACTIVITY_TYPE_CD', 'PAYMENT_NUM', 'ROAD_NM_CD', 'LOTNO_CD',
    'PAYMENT_AMT_WON', 'PAYMENT_ETC', 'SGG_CD', 'REL_CD', 'COMPANION_SITUATION', 'LODGING_TYPE_CD',
    'RSVT_YN', 'PAYMENT_NUM_lodge', 'ROAD_NM_CD_lodge', 'LOTNO_CD_lodge', 'PAYMENT_AMT_WON_lodge',
    'PAYMENT_ETC_lodge', 'MVMN_SE', 'PAYMENT_SE', 'MVMN_SE_NM', 'PAYMENT_NUM_mvmn', 'PAYMENT_AMT_WON_mvmn',
    'PAYMENT_ETC_mvmn', 'TRAVELER_ID', 'RESIDENCE_SGG_CD', 'GENDER', 'AGE_GRP', 'INCOME',
    'TRAVEL_STATUS_RESIDENCE', 'TRAVEL_STATUS_DESTINATION', 'TRAVEL_STATUS_ACCOMPANY',
    'TRAVEL_COMPANIONS_NUM', 'TRAVEL_PURPOSE', 'TRAVEL_START_YMD', 'TRAVEL_END_YMD', 'MVMN_NM', 'TRAVEL_DAY'
]

# 병합된 데이터프레임(merged_df)에서 필요한 열만 선택합니다.
filtered_df = merged_df[selected_columns]

# 선택한 열들로 구성된 데이터를 CSV 파일로 저장합니다.
filtered_df.to_csv('filtered_travel_data.csv', index=False, encoding='utf-8')
```
